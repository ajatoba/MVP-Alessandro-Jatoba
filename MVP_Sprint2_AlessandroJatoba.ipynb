```
# Configuração para não exibir os warnings
import warnings
warnings.filterwarnings("ignore")

# Imports necessários
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split # para particionar em bases de treino e teste (holdout)
from sklearn.model_selection import KFold # para preparar os folds da validação cruzada
from sklearn.model_selection import cross_val_score # para executar a validação cruzada
from sklearn.metrics import mean_squared_error # métrica de avaliação MSE
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LinearRegression # algoritmo Regressão Linear
from sklearn.neighbors import KNeighborsRegressor # algoritmo KNN
from sklearn.tree import DecisionTreeRegressor # algoritmo Árvore de Regressão
from sklearn.svm import SVR # algoritmo SVM
from sklearn.pipeline import Pipeline
from sklearn.ensemble import ExtraTreesRegressor #importando o algoritmo para Feature Selection
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import VotingRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```
# Carga do dataset
dataset = pd.DataFrame(pd.read_csv("https://raw.githubusercontent.com/ajatoba/MVP-Alessandro-Jatoba/main/dataset_health_system.csv", sep = ';')) # carga do arquivo CSV e conversão para dataframe

dataset['target'] = dataset.target # definição da coluna target

dataset.head()
```

**Tratando variáveis dummy**

O dataset tem muitas variáveis dummy. É necessário tratá-las para que não haja interferência nos resultados

```
dummies = ['Updated_health_workforce_strategy_to_address_human_resource_shortfalls',
            'In-country_capacity_to_isolate_patients_with_highly_communicable_diseases',
            'Constitutional_guarantee_of_citizens_right_to_medical_care',
            'Demonstrated_capacity___evidence_of_plan_to_expand_isolation_capacity',
            'National_procurement_protocol_for_the_acquisition_of_routine_laboratory_medical_supplies',
            'Stockpile_of_medical_supplies_for_national_use_during_a_public_health_emergency',
            'Stockpile_of_laboratory_supplies_for_national_use_during_a_public_health_emergency',
            'Annual_review_of_national_stockpile_to_ensure_sufficient_supply',
            'Plan_agreement_to_produce_procure_medical_supplies_during_a_public_health_emergency',
            'Plan_agreement_to_produce_procure_lab_supplies_during_a_public_health_emergency',
            'Plan_program_guidelines_for_dispensing_MCMs_during_a_public_health_emergency',
            'Plan_to_receive_foreign_health_personnel_during_a_public_health_emergency',
            'Guaranteed_paid_sick_leave',
            'Government_prioritisation_of_care_for_healthcare_workers_during_response',
            'Existence_of_system_for_communication_during_a_public_health_emergency_',
            'Inclusion_of_public_and_private_sector_in_healthcare_communication_system_',
            'Evidence_of_national_public_health_system_monitoring_and_tracking_of_HCAIs',
            'Requirement_for_ethical_review_before_beginning_a_clinical_trial',
            'Expedited_approval_for_clinical_trials_of_unregistered_MCMs_during_epidemics_',
            'Existence_of_agency_responsible_for_approving_new_human_MCMs',
            'Expedited_approval_for_human_MCMs_during_public_health_emergencies_']

dataset = pd.get_dummies(dataset, columns=dummies, drop_first=True)
dataset.info()
dataset.head()
```

```
# Preparação dos dados
test_size = 0.20 # tamanho do conjunto de teste
seed = 7
n_estimators=100
max_features = 3

# Separação em bases de treino e teste (holdout)
array = dataset.values
X = array[:,1:] # atributos (pegando a partir do índice 1, visto que o índice 0 é o target)
y = dataset.target# classe (target)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=seed) # faz a divisão

# Parâmetros e partições da validação cruzada
scoring = 'accuracy'
num_particoes = 10
kfold = KFold(n_splits=num_particoes, shuffle=True, random_state=seed) # validação cruzada sem estratificação
```

**Feature selection**

Escolhido fazer a seleção de atributos usando algoritmo ExtraTrees, tendo em vista a grande quantidade de atributos do dataset, para não ter que escolher um número de atributos importantes.

```
modelo = ExtraTreesRegressor(n_estimators=n_estimators)
modelo.fit(X,y)

print("\nAtributos originais:", dataset.columns[1:33])# Exibindo os atributos originais
print(modelo.feature_importances_) # Exibindo os scores de cada atributo
```

**Normalização**

A maioria das variáveis do dataset é categórica, no entanto algumas são valores, em especial, porcentagens. Optei, portanto, por normalizar e padronizar essas variáveis.

```
# Modelagem

# Definindo uma seed global para esta célula de código
np.random.seed(seed)

# Listas para armazenar os armazenar os pipelines e os resultados para todas as visões do dataset
pipelines = []
results = []
names = []

# Algoritmos que serão utilizados
reg_lin = ('LR', LinearRegression())
knn = ('KNN', KNeighborsRegressor())
cart = ('CART', DecisionTreeRegressor())
svm = ('SVM', SVR())
extra_trees = ('ET', ExtraTreesRegressor(n_estimators=n_estimators, max_features=max_features))

# Transformações que serão utilizadas
standard_scaler = ('StandardScaler', StandardScaler())
min_max_scaler = ('MinMaxScaler', MinMaxScaler())

# Dataset original
pipelines.append(('LR-orig', Pipeline([reg_lin])))
pipelines.append(('KNN-orig', Pipeline([knn])))
pipelines.append(('CART-orig', Pipeline([cart])))
pipelines.append(('SVM-orig', Pipeline([svm])))
pipelines.append(('ET-orig', Pipeline([extra_trees])))

# Dataset Padronizado
pipelines.append(('LR-padr', Pipeline([standard_scaler, reg_lin])))
pipelines.append(('KNN-padr', Pipeline([standard_scaler, knn])))
pipelines.append(('CART-padr', Pipeline([standard_scaler, cart])))
pipelines.append(('SVM-padr', Pipeline([standard_scaler, svm])))
pipelines.append(('ET-padr', Pipeline([standard_scaler, extra_trees])))

# Dataset Normalizado
pipelines.append(('LR-norm', Pipeline([min_max_scaler, reg_lin])))
pipelines.append(('KNN-norm', Pipeline([min_max_scaler, knn])))
pipelines.append(('CART-norm', Pipeline([min_max_scaler, cart])))
pipelines.append(('SVM-norm', Pipeline([min_max_scaler, svm])))
pipelines.append(('ET-norm', Pipeline([min_max_scaler, extra_trees])))

# Execução das comparações
for name, model in pipelines:
  cv_results = cross_val_score(model, X_train, y_train, cv=kfold, scoring='neg_mean_squared_error')
  results.append(cv_results)
  names.append(name)
  msg = "%s: %.3f (%.3f)" % (name, cv_results.mean(), cv_results.std()) # formatando para 3 casas decimais
  print(msg)

# Boxplot de comparação dos modelos
fig = plt.figure(figsize=(25,6))
fig.suptitle('Comparação dos Modelos - Dataset orginal, padronizado e normalizado')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()#
```


O algoritmo de Regressão Linear padronizado foi o modelo de melhor desempenho
Não há hiperparâmetros para serem ajustados ou otimizados, além de não terem sido observados problemas de underfiting

O modelo, adotando o algoritmo de melhor desempenho, será treinado e suas predições com o conjunto de treino avaliadas


```
# Criando um modelo com todo o conjunto de treino
model = LinearRegression()#Regressão Linear foi o modelo de melhor desempenho

model.fit(X_train, y_train)

# Fazendo as predições com o conjunto de treino
predictions = model.predict(X_train)

# Estimando o MSE e o RMSE no conjunto de treino
mse = mean_squared_error(y_train, predictions)

# Exibindo os resultados
print("MSE %0.2f" % mse)
print("RMSE %0.2f" % np.sqrt(abs(mse)))
```

```
 #Avaliação do modelo com o conjunto de testes

# Preparação do modelo
scaler = StandardScaler().fit(X_train) # Padronização com o conjunto de treino
rescaledX = scaler.transform(X_train) # Aplicação da padronização no conjunto de treino
model.fit(rescaledX, y_train)

# Estimativa da acurácia no conjunto de teste
rescaledTestX = scaler.transform(X_test) # aplicação da padronização no conjunto de teste

# Fazendo as predições com o conjunto de teste
predictions = model.predict(rescaledTestX)

# Exibindo os resultados
print(mean_squared_error(y_test, predictions))
```
